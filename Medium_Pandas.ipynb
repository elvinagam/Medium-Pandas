## Querying

```python
#@title First version
df.loc[(df['loading'] > 100) & (df['loading'] < 110)]
```

```python
#@title Medium version
df.query('loading > 100 and loading < 110')
# alternatively, use f string method for repetability
min_loading = 100
max_loading = 110
df.query(f'loading > {min_loading} and loading < {max_loading}')
# alternatively, simply use @ symbol to access variables
df.query('loading > @min_loading and loading <@max_loading')
```

## Vectorization: Iterate over rows

```python
#@title First version
for i , row in df.iterrows():
  if row['attribute_1'] == 'material_8':
    df.loc[i, 'is_material'] = True
  else:
    df.loc[i, 'is_material'] = False
df[['is_material']]
```

```python
#@title Medium version
df['is_material'] = df['attribute_1'] == 'material_8'
df[['is_material']]
```

## Vectorization: Lambda functions

```python
#@title First version
df['loading_squared'] = df.apply(lambda row: row['loading']**2, axis=1)
df[['loading_squared']]
```

```python
#@title Medium version
df['loading_squared'] = df['loading'] ** 2
df[['loading_squared']]
```

## Aggregating throuh columns

```python
#@title First version
df['change'] = df['loading'].diff()
df['pct_change'] = df['loading'].pct_change()

df
```
